#### Formas de mudar a paleta de cores 

##### 1Âª forma

```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr']
producao_1 = [500, 700, 800, 650]
producao_2 = [480, 720, 750, 700]

cores = plt.get_cmap('Dark2').colors

plt.plot(meses, producao_1, label='Fazenda A', marker='o', color=cores[0])
plt.plot(meses, producao_2, label='Fazenda B', linestyle='--', marker='s', color=cores[1])

plt.title("ProduÃ§Ã£o Mensal de Peixes")
plt.xlabel("MÃªs")
plt.ylabel("ProduÃ§Ã£o (kg)")

#plt.legend(loc='upper left', fontsize=10, shadow=True)  # Legenda com sombra e borda
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

meses = ['Jan', 'Fev', 'Mar', 'Abr']
producao_1 = [500, 700, 800, 650]
producao_2 = [480, 720, 750, 700]

plt.rc('axes', prop_cycle=(cycler('color', plt.get_cmap('Set1').colors)))

plt.plot(meses, producao_1, label='Fazenda A', marker='o')
plt.plot(meses, producao_2, label='Fazenda B', linestyle='--', marker='s')

plt.title("ProduÃ§Ã£o Mensal de Peixes")
plt.xlabel("MÃªs")
plt.ylabel("ProduÃ§Ã£o (kg)")

#plt.legend(loc='upper left', fontsize=10, shadow=True)  # Legenda com sombra e borda
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

meses = ['Jan', 'Fev', 'Mar', 'Abr']
producao_1 = [500, 700, 800, 650]
producao_2 = [480, 720, 750, 700]

cores = plt.get_cmap('Dark2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

plt.plot(meses, producao_1, label='Fazenda A', marker='o')
plt.plot(meses, producao_2, label='Fazenda B', linestyle='--', marker='s')

plt.title("ProduÃ§Ã£o Mensal de Peixes")
plt.xlabel("MÃªs")
plt.ylabel("ProduÃ§Ã£o (kg)")

#plt.legend(loc='upper left', fontsize=10, shadow=True)  # Legenda com sombra e borda
plt.grid(True)
plt.tight_layout()
plt.show()
```
