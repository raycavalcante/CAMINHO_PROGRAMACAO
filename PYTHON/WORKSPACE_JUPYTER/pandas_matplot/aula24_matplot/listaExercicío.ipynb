```python
import matplotlib.pyplot as ray
```

```python
# ques01
dias = [1,2,3,4,5,6,7]
temperaturas = [30,32,33,31,29,28,27]
# craindo o grafico
ray.plot(dias,temperaturas,  color='red', marker='D')
# ray.bar(dias,temperaturas,  color='green')
# barras = ray.bar(dias,temperaturas,  color='red')
# ray.bar_label(barras, labels=temperaturas, padding=1, fontsize=10, fontweight='bold')
# ray.scatter(dias,temperaturas, color='red')
ray.xlabel('DIAS')
ray.ylabel('TEMPERATURA')
ray.title('TEMPERATURA POR DIA')
# ray.grid(True)
ray.show()
```

```python
# ques02
dias = ['Segunda', 'Tera', 'Quata', 'Quinta', 'Sexta']
vendas = [150, 200, 170, 240, 300]

barras = ray.bar(dias, vendas, color='blue')
ray.bar_label(barras, labels=vendas, padding=3, fontsize=8, fontweight='bold')
ray.xlabel('DIAS')
ray.ylabel('VENDAS')
ray.title('VENDAS DURANTE O MS')
ray.show()
```

```python
# ques03
anos = [2000, 2005, 2010, 2015, 2020]
populacao = [1000, 1200, 1350, 1600, 1800]

ray.plot(anos,populacao, color='blue', marker='d')
ray.xlabel('ANO')
ray.ylabel('POPULAO (MILHARES)')
ray.title('CRESCIMENTO POPULACIONAL')
ray.grid(True)
ray.show()
```

```python
# ques04
produtos = ['Arroz', 'Feijo', 'Farinha', 'Aai']
quantidades = [30, 50, 45, 70]
destacar = [0,0,0,0.1]

# ray.pie(quantidades)
# partes = ray.pie(quantidades, labels=produtos)
# partes = ray.pie(quantidades, labels=produtos, autopct='%.1f')
partes = ray.pie(quantidades, labels=produtos, autopct='%.1f',explode=destacar)


ray.title('DISTRIBUIO DOS PRODUTOS')
ray.show()
```

```python
#ques05
dias = ['Segunda', 'Tera', 'Quata', 'Quinta', 'Sexta','sabado','domingo']
tefe = [28, 29, 30, 31, 32, 33, 34]
tonantins = [25, 26, 27, 28, 29, 28, 27]

ray.plot(dias, tefe,label='Tef', marker='o')
ray.plot(dias, tonantins,label='Tonantins', linestyle='--',  marker='o')
ray.xlabel('Dias')
ray.ylabel('Temperaturas')
ray.title("Temperatura das cidade de Tef e Tonantins")
ray.grid(True)
ray.legend()
ray.show()
```
