```python
import matplotlib.pyplot as plt

# Dados de exemplo
meses = ['Jan', 'Fev', 'Mar', 'Abr']
vendas_loja1 = [1500, 1800, 1200, 2000]
vendas_loja2 = [1600, 1750, 1300, 1900]

# Criar uma figura com 1 linha e 2 colunas de gráficos
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Primeiro gráfico (Loja 1)
axs[0].bar(meses, vendas_loja1, color='skyblue')
axs[0].set_title("Vendas - Loja 1")
axs[0].set_ylabel("R$")

# Segundo gráfico (Loja 2)
axs[1].bar(meses, vendas_loja2, color='salmon')
axs[1].set_title("Vendas - Loja 2")
axs[1].set_ylabel("R$")

# Ajustar o espaçamento
plt.tight_layout()

plt.savefig("grafico_vendas.pdf", dpi=300, bbox_inches='tight') # Salvar o gráfico

# Mostrar ambos os gráficos
plt.show()
```

```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr']
producao_1 = [500, 700, 800, 650]
producao_2 = [480, 720, 750, 700]

plt.plot(meses, producao_1, label='Fazenda A', color='green', marker='o')
plt.plot(meses, producao_2, label='Fazenda B', color='orange', linestyle='--', marker='s')

plt.title("Produção Mensal de Peixes")
plt.xlabel("Mês")
plt.ylabel("Produção (kg)")

plt.legend(loc='upper left', fontsize=10, frameon=True, shadow=True)  # Legenda com sombra e borda
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
producao = [500, 600, 800, 750, 900, 950]  # em kg
temperatura = [28, 29, 30, 31, 30, 29]     # em °C

fig, ax1 = plt.subplots()

# Primeiro eixo (produção)
ax1.set_xlabel("Mês")
ax1.set_ylabel("Produção (kg)", color='blue')
ax1.plot(meses, producao, color='blue', marker='o')
ax1.tick_params(axis='y', labelcolor='blue')

# Segundo eixo (temperatura)
ax2 = ax1.twinx()  # Cria um segundo eixo Y, compartilhando o mesmo eixo X
ax2.set_ylabel("Temperatura (°C)", color='red')
ax2.plot(meses, temperatura, color='red', linestyle='--', marker='s')
ax2.tick_params(axis='y', labelcolor='red')

plt.title("Produção vs Temperatura")
plt.grid(True)
plt.tight_layout()
plt.show()
```
