```python
vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")

vendas_df = pd.concat([vendas_dez_df,vendas_df])
display(vendas_df)
```

```python
# vendas_lojas = vendas_df['Produto'].value_counts()
# display(vendas_lojas)
produto = vendas_df[['Produto','Valor Final']].groupby('Produto').sum()
display(produto)
```
